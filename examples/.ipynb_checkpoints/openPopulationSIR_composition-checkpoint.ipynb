```julia
using AlgebraicStockFlow

using LabelledArrays
using OrdinaryDiffEq
using Plots
using JSON

using Catlab
using Catlab.CategoricalAlgebra
using Catlab.Graphics
using Catlab.Programs
using Catlab.Theories
using Catlab.WiringDiagrams

using LabelledArrays
using LinearAlgebra: mul!

using Catlab.WiringDiagrams
using Catlab.Programs.RelationalPrograms
```

```julia
# help functions
state_dict(n) = Dict(s=>i for (i, (s,v)) in enumerate(n))
parav(pa,ps)=pa[state_dict(pa)[ps]][2]
final_state(m)=[(y,x) for (x,y) in m.tables.S]
```

# SIR model

```julia
st_sir=[]
para_sir=[(:birthRate,0.001), (:c_beta,0.1), (:recRate, 0.05), (:deathRate, 0.001)]
```

# NOTE: the functions maybe need to be modified after composition. For example, if we need total population N(t), if there more stocks adds to the system after composition, then the part of N(t) need to be modified

## here, in the original f_birth and f_inf: N = S+I+R 

```julia
f_birth(u,t)=parav(para_sir,:birthRate)*(u[state_dict(st_sir)[:S]]+u[state_dict(st_sir)[:I]]+u[state_dict(st_sir)[:R]])
# here use "*rand((1,2,3,4,5))" to test the solution of the stochastic models. And the results shows the flows would be 
# evaluated multi times instead of once if we use the OrdinaryDiffEq pkg. So, I think we need to implement the solution
# method by ourself, where make sure each flow should only be evaluated once!!
f_inf(u,t)= parav(para_sir,:c_beta)*u[state_dict(st_sir)[:S]]*u[state_dict(st_sir)[:I]]/(u[state_dict(st_sir)[:S]]+u[state_dict(st_sir)[:I]]+u[state_dict(st_sir)[:R]])#*rand((1,2,3,4,5))
f_rec(u,t)= parav(para_sir,:recRate)*u[state_dict(st_sir)[:I]]
f_deathS(u,t)=u[state_dict(st_sir)[:S]]*parav(para_sir,:deathRate)
f_deathI(u,t)=u[state_dict(st_sir)[:I]]*parav(para_sir,:deathRate)
f_deathR(u,t)=u[state_dict(st_sir)[:R]]*parav(para_sir,:deathRate)
```

```julia
sir_LabelledStockFlow=LabelledStockFlow{Function, Int}((:birth=>f_birth, :inf=>f_inf, :rec=>f_rec, :deathS=>f_deathS, :deathI=>f_deathI, :deathR=>f_deathR), (:S,990)=>(:birth => (:inf,:deathS)), (:I,10)=>(:inf => (:rec,:deathI)),(:R,0)=>(:rec=>:deathR))
```

```julia
Graph(sir_LabelledStockFlow)
```

```julia
st_sir=final_state(sir_LabelledStockFlow)
prob = ODEProblem(vectorfield(sir_LabelledStockFlow),initialValues(sir_LabelledStockFlow),(0.0,100.0),funcFlows(sir_LabelledStockFlow));
sol = solve(prob,Tsit5(),abstol=1e-8);
plot(sol)
```

# SV

```julia
st_sv=[]
para_sv=[(:vaccRate,0.1)]
f_vacc(u,t)=u[state_dict(st_sv)[:S]]*parav(para_sv,:vaccRate)
f_deathV(u,t)=u[state_dict(st_sv)[:V]]*parav(para_sir,:deathRate)
```

```julia
sv_LabelledStockFlow=LabelledStockFlow{Function, Int}((:vacc=>f_vacc, :deathV=>f_deathV), (:S,990)=>((:F_NONE,:F_NONE) => :vacc), (:V,0)=>(:vacc => :deathV))
```

```julia
Graph(sv_LabelledStockFlow)
```

```julia
st_sv=final_state(sv_LabelledStockFlow)
prob = ODEProblem(vectorfield(sv_LabelledStockFlow),initialValues(sv_LabelledStockFlow),(0.0,100.0),funcFlows(sv_LabelledStockFlow));
sol = solve(prob,Tsit5(),abstol=1e-8);
plot(sol)
```

# compose SIR and SV

```julia
sirv = @relation (V,S,I,R) begin
  sv(S,V)
  sir(S,I,R) 
end
```

```julia
sir_sc=Open(sir_LabelledStockFlow)
sv_sc=Open(sv_LabelledStockFlow)
sv_dict = Dict(:sv=>sv_sc,:sir=>sir_sc);
sirv_stockflow = apex(oapply(sirv, sv_dict))
```

```julia
Graph(sirv_stockflow)
```

## Redefine f_birth, since now, the N=S+I+R+V
## this means, after compostion, we need to redefine some flows' functions if they would be changed due to composition

```julia
f_birth(u,t)=parav(para_sir,:birthRate)*(u[state_dict(st_sir)[:S]]+u[state_dict(st_sir)[:I]]+u[state_dict(st_sir)[:R]]+u[state_dict(st_sir)[:V]])
f_inf(u,t)= parav(para_sir,:c_beta)*u[state_dict(st_sir)[:S]]*u[state_dict(st_sir)[:I]]/(u[state_dict(st_sir)[:S]]+u[state_dict(st_sir)[:I]]+u[state_dict(st_sir)[:R]]+u[state_dict(st_sir)[:V]])#*rand((1,2,3,4,5))
```

```julia
# now set the state for the user defined functions
st_sv=final_state(sirv_stockflow)
st_sir=final_state(sirv_stockflow)
prob = ODEProblem(vectorfield(sirv_stockflow),initialValues(sirv_stockflow),(0.0,100.0),funcFlows(sirv_stockflow));
sol = solve(prob,Tsit5(),abstol=1e-8);
plot(sol)
```

```julia
sum(sol[24])
```

# if the V is not 100% protected, then, we can have a flow V->I. Let's do it by composition

```julia
st_vi=[]
para_vi=[(:vaccEffi,0.9)]
f_infv(u,t)=u[state_dict(st_vi)[:V]]*(1-parav(para_vi,:vaccEffi))*u[state_dict(st_vi)[:I]]/(u[state_dict(st_sir)[:S]]+u[state_dict(st_sir)[:I]]+u[state_dict(st_sir)[:R]]+u[state_dict(st_sir)[:V]])
```

```julia
vi_LabelledStockFlow=LabelledStockFlow{Function, Int}((:infv=>f_infv), (:V,0)=>(:F_NONE => :infv), (:I,10)=>(:infv => :F_NONE))
```
